```python
# Function to embed a set of tokens and reshape the embedding tensor
def get_embeddings_and_reshape(model, token_set, embedding_dim=64):
    embeddings = model.embed(token_set)
    return embeddings.view(len(token_set), embedding_dim)

# Function to calculate vector norms
def calculate_vector_norms(embeddings):
    return torch.norm(embeddings, dim=1)

# Function to plot histograms of vector norms
def plot_vector_norm_histogram(vector_norms, title):
    plt.hist(vector_norms.detach().numpy(), bins=50)
    plt.xlabel('Vector Lengths')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.show()
```

#Making linear probes

```python
class LinearProbe(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(LinearProbe, self).__init__()
        self.linear = nn.Linear(input_dim, output_dim)

    def forward(self, x):
        return self.linear(x)
```

```python
class ProbeTrainerModel(nn.Module):
    def __init__(self):
        super(ProbeTrainerModel, self).__init__()
        self.hooked_model = HookedTransformer.from_pretrained('roneneldan/TinyStories-1M')
        for param in self.hooked_model.parameters():
            param.requires_grad = False
        self.classifiers = nn.ModuleList([LinearProbe(64, 4) for _ in range(8)])
        self.other_class_weights = torch.mean(avg, dim=0).unsqueeze(0)  # or torch.zeros(1, 64)

        for classifier in self.classifiers:
            # Copy over the weights for the noun, verb, and adjective classes
            classifier.linear.weight.data[:3, :] = avg
            # Set the weights for the "other" class
            classifier.linear.weight.data[3, :] = -self.other_class_weights

    def forward(self, input_string, token_pos):
        outputs, cache = self.hooked_model.run_with_cache(input_string)
        layers = [f"blocks.{i}.hook_resid_pre" for i in range(8)]
        batch_size = cache[layers[0]].shape[0]
        ans = torch.zeros((8, batch_size, 4)).to(device)  # ensure the tensor is on the correct device
        for idx in range(8):
            selected_outputs = torch.gather(cache[layers[idx]], 1, token_pos.unsqueeze(1).unsqueeze(2).expand(-1, -1, 64))
            ans[idx] = self.classifiers[idx](selected_outputs.squeeze(1))

        ans = ans.sum(dim=0)  # Sum over the first dimension

        return ans
```

```python
class PartOfSpeechDataset(Dataset):
    def __init__(self, datasets, labels):
        self.data = []
        for dataset, label in zip(datasets, labels):
            for item in dataset:
                self.data.append({
                    'input_ids': item['input_ids'],
                    'token_pos': item['token_pos'],
                    'label': label  # label is a scalar representing the part of speech
                })

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        input_ids = item['input_ids']
        token_pos = item['token_pos']
        label = item['label']

        # Calculate the window in which we can shift the start of our slice
        # while still including the token of interest
        window_start = max(0, token_pos - 19)  # If token_pos is less than 19, we can't shift the start
        window_end = min(len(input_ids) - 20, token_pos)  # If token_pos is too close to the end, we can't shift the start

        # Randomly select the start of our slice from the window
        start_pos = random.randint(window_start, window_end)
        end_pos = start_pos + 20

        shifted_input_ids = input_ids[start_pos : end_pos]
        shifted_token_pos = token_pos - start_pos

        return {
            'input_ids': shifted_input_ids,
            'token_pos': shifted_token_pos,
            'label': label
        }
```

```python
def find_words_with_tokens(data_list, token_ids_set):
    '''Returns a dictionary mapping each token to its tokenized story and position'''
    token_positions = {token_id: None for token_id in token_ids_set}
    for story_idx, item in enumerate(data_list):
        text = item['text']
        tokenized_text = tokenizer.encode(text, return_tensors='pt')[0]
        for token_id in token_ids_set:
            try:
                token_pos = (tokenized_text == token_id).nonzero(as_tuple=True)[0].item()
                token_positions[token_id] = (tokenized_text, token_pos)
            except ValueError:
                continue  # Token not found in this story, move to the next
    return token_positions

class WordPositionDataset(Dataset):
    def __init__(self, token_positions_dict, context_window=40):
        self.data = []
        self.context_window = context_window
        for token_id, (tokenized_text, token_pos) in token_positions_dict.items():
            if tokenized_text is not None and token_pos is not None:
                # Compute start and end indices for context window
                half_window = self.context_window // 2
                start_idx = max(token_pos - half_window, 0)
                end_idx = start_idx + self.context_window
                if end_idx > len(tokenized_text):
                    end_idx = len(tokenized_text)
                    start_idx = max(end_idx - self.context_window, 0)
                self.data.append({
                    'input_ids': tokenized_text[start_idx:end_idx],
                    'token_pos': token_pos - start_idx,
                    'label': torch.tensor(token_id)  # Using token_id as label for simplicity
                })

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx]
```

```python
def patch_residual_stream(activations, hook, tensors, intervention_details):
    # The intervention details will be a list of tuples: (token_idx, layer_idx, direction_idx, magnitude)
    for token_idx, layer_idx, direction_idx, magnitude in intervention_details:
        # Make sure that token_idx is within the range of the activations shape
        if token_idx < activations.shape[1]:
            # Apply the intervention at the specific layer, direction, and magnitude
            activations[:, token_idx, :] += tensors[layer_idx, direction_idx, :] * magnitude
    return activations
def add_intervention_hooks(model, tensors, intervention_details):
    # Sort intervention details by layer, as we need to add hooks layer by layer
    intervention_details.sort(key=lambda x: x[1])
    # Group interventions by layer
    interventions_by_layer = {}
    for token_idx, layer_idx, direction_idx, magnitude in intervention_details:
        if layer_idx not in interventions_by_layer:
            interventions_by_layer[layer_idx] = []
        interventions_by_layer[layer_idx].append((token_idx, direction_idx, magnitude))

    # Add hooks for each layer where interventions are needed
    for layer_idx, interventions in interventions_by_layer.items():
        # Define the partial function for patching
        patch_fn = partial(patch_residual_stream, tensors=tensors, intervention_details=interventions)
        # Add the hook to the model
        hook_name = f'blocks.{layer_idx}.hook_resid_post'
        model.add_hook(hook_name, patch_fn)
```

```python
def patch_residual_stream(activations, hook, layer, pos, adjustment_vector):
    # Adjust the residual stream at the specified position with the given vector
    activations[layer, pos, :] += adjustment_vector
    return activations

def generate_with_intervention(model, initial_string, x, adj_subtract_old, noun_add_old, noun_subtract_man, verb_add_man, tensors):
    generated_string = initial_string
    for _ in range(x):
        # Tokenize the current string to get a list of tokens
        tokenized_string = model.tokenizer.encode(generated_string)

        # Identify the positions of "old" and "man" in the tokenized string
        pos_old = tokenized_string.index(model.tokenizer.encode("old")[0])
        pos_man = tokenized_string.index(model.tokenizer.encode("man")[0])

        # Calculate the adjustment vectors for "old" and "man"
        adj_vector_old = tensors[:, 2, :] * adj_subtract_old + tensors[:, 0, :] * noun_add_old
        noun_vector_man = tensors[:, 0, :] * -noun_subtract_man + tensors[:, 1, :] * verb_add_man

        # Apply the intervention using hooks
        model.add_hook(layer='blocks.6.hook_resid_post',
                       hook=partial(patch_residual_stream, layer=6, pos=pos_old, adjustment_vector=adj_vector_old))
        model.add_hook(layer='blocks.6.hook_resid_post',
                       hook=partial(patch_residual_stream, layer=6, pos=pos_man, adjustment_vector=noun_vector_man))

        # Generate logits using the model for the tokenized string with interventions
        logits, _ = model.run_with_cache(generated_string)

        # Remove the hooks after running the model to avoid affecting subsequent calls
        model.reset_hooks()

        # Get the logits for the next token (the last token in the sequence)
        next_token_logits = logits[0, -1, :]
        next_token = next_token_logits.argmax(dim=-1).item()

        # Decode the next token and add it to the generated string
        generated_string += model.tokenizer.decode([next_token])

    return generated_string[len(initial_string):]  # Return only the newly generated part

# Parameters for intervention
adj_subtract_old = 0.5  # Example value, adjust as needed
noun_add_old = 0.5      # Example value, adjust as needed
noun_subtract_man = 0.5 # Example value, adjust as needed
verb_add_man = 0.5      # Example value, adjust as needed

# Test the function with intervention
initial_string = "the old man"
x = 5
generated = generate_with_intervention(model, initial_string, x, adj_subtract_old, noun_add_old, noun_subtract_man, verb_add_man, tensors)
print("Generated with intervention:", generated)
```
